```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10, 8]

for index, text in enumerate(data_clean.index):
    x = data_clean.polarity.loc[text]
    y = data_clean.subjectivity.loc[text]
    plt.scatter(x, y, color='blue')
    
    plt.xlim(-1, 1) 
    
plt.title('Sentiment Analysis', fontsize=20)
plt.xlabel('<-- Negative -------- Positive -->', fontsize=15)
plt.ylabel('<-- Facts -------- Opinions -->', fontsize=15)

plt.show()
```

```python
import datetime
import matplotlib.pyplot as plt
from matplotlib.dates import DayLocator, HourLocator, DateFormatter, drange
import numpy as np

dates=data_clean.date[:6]
y=data_clean.polarity[:6]
fig, ax = plt.subplots()
ax.plot_date(dates, y , 'g')
ax.set_xlim(dates[1], dates[5])

ax.xaxis.set_major_locator(DayLocator())
ax.xaxis.set_minor_locator(HourLocator(range(1, 26, 6)))
#ax.xaxis.set_major_formatter(DateFormatter('% Y-% m-% d'))

ax.fmt_xdata = DateFormatter('% Y-% m-% d H:% M:% S')
fig.autofmt_xdate()
plt.title('Sentiment over time', fontweight ="bold")
plt.show()
```
